# 🟫 Bronze Layer – Raw Data Ingestion

The Bronze layer contains raw, unprocessed data exactly as it arrives from the source systems.  
Its purpose is to preserve the original data for traceability, auditing, and replay.

### Key characteristics
- Data is stored in its **original format** (CSV, JSON, Parquet, etc.)
- No transformations are applied
- Schema may be semi‑structured or unstructured
- Used as the foundation for Silver and Gold layers

### Typical tasks in this layer
- Ingest data from external sources (API, cloud storage, databases)
- Apply minimal metadata (ingestion timestamp, source file name)
- Store data in Delta format for reliability and versioning



Create Schema

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Bronze layer - raw ingested data';
```

Inspect Source Data

```python
%sql
SELECT *
FROM databricks_simulated_retail_customer_data.v01.sales_orders
LIMIT 10;
```

Create table bronze.sales_orders

```python
%sql
CREATE OR REPLACE TABLE bronze.sales_orders
USING DELTA
AS
SELECT *
FROM databricks_simulated_retail_customer_data.v01.sales_orders;
```

Add Ingestion Timestamp

```python
%sql
CREATE OR REPLACE TABLE bronze.sales_orders AS
SELECT
    *,
    current_timestamp() AS ingestion_timestamp
FROM databricks_simulated_retail_customer_data.v01.sales_orders;
```

```python
%sql
SELECT COUNT(*) FROM bronze.sales_orders;
```

```python
%sql
SELECT * FROM bronze.sales_orders LIMIT 10;
```

```python
%sql
DESCRIBE TABLE bronze.sales_orders;
```

Final Output

```python
%sql
DESCRIBE TABLE bronze.sales_orders;

SELECT *
FROM bronze.sales_orders
LIMIT 10;
```
